```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import csv
```

```python
req = requests.get("https://www.flipkart.com/search?q=mobiles&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&as-pos=1&as-type=RECENT&suggestionId=mobiles%7CMobiles&requestId=9f422b53-af60-41c4-bf77-4b3a52a66798&as-searchtext=mobile")
content = req.content
```

```python
soup = BeautifulSoup(content,'html.parser')
soup.prettify()
```

```python
price=soup.find_all('div',class_='_30jeq3 _1_WHN1')
price
```

```python
prices=[]
for i in price:
    prices.append(i.text)
print(prices)
print(len(prices))
```

```python
nm=soup.find_all('div',class_='_4rR01T')
```

```python
names=[]
for i in nm:
    names.append(i.text)
print(names)
print(len(names))
```

```python
cmn_class=soup.find_all('li',class_='rgWa7D')
```

```python
cmon_cls=[]
for i in cmn_class:
    cmon_cls.append(i.text)
print(len(cmon_cls))
```

```python
ram=[]
processor=[]

battery=[]
for i in range(len(cmon_cls)):
    p=cmon_cls[i]
    if 'RAM' in p:
        ram.append(p)
    elif 'Processor' in p:
        processor.append(p)
    elif 'Battery' in p:
        battery.append(p)

               
               
len(processor)         
               
               
    
```

```python
df={'Description':names,'Processor':processor,'Price':prices}
```
